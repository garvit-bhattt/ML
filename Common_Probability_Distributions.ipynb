# 🎲 Topic 2: Common Probability Distributions

Think of distributions as **mathematical templates**. Instead of calculating probabilities from scratch every time, we realize that data in the real world tends to follow a few standard shapes. In this guide, we'll cover the four building blocks of probability distributions.

## 1. 🧠 Intuition First

*   **Bernoulli:** The simplest of them all. A single event with two outcomes: **Success (1)** or **Failure (0)**. Think of a single coin flip.
*   **Binomial:** What happens if you do a Bernoulli trial multiple times? If you flip a coin 10 times, what's the chance of getting exactly 7 heads? That’s Binomial.
*   **Exponential:** The "waiting time" distribution. How long until the next earthquake? How long until the bus arrives? It models the **time between events**.
*   **Gaussian (Normal):** The king of distributions. The classic **Bell Curve**. Why is it everywhere? Because in nature, when you add up millions of tiny random variables (like genetics and nutrition determining human height), they naturally smooth out into a Gaussian shape.

## 2. 📌 Key Concepts (Parameters & Moments)

**Parameters** are the "knobs" you turn to change the shape of the distribution. **Moments** are summaries of the shape, primarily the **Mean** (Expected Value, or center) and **Variance** (Spread/Width).

### **Bernoulli($p$)**
*   **Parameter:** $p$ (probability of success).
*   **Mean:** $p$
*   **Variance:** $p(1-p)$

### **Binomial($n, p$)**
*   **Parameters:** $n$ (number of trials), $p$ (probability of success per trial).
*   **Mean:** $np$
*   **Variance:** $np(1-p)$

### **Exponential($\lambda$)**
*   **Parameter:** $\lambda$ (rate of events, e.g., 5 buses per hour).
*   **Mean:** $1/\lambda$ (wait time, e.g., 1/5 hour)
*   **Variance:** $1/\lambda^2$

### **Gaussian/Normal($\mu, \sigma^2$)**
*   **Parameters:** $\mu$ (mean), $\sigma^2$ (variance). Standard deviation is $\sigma$.
*   **Mean:** $\mu$
*   **Variance:** $\sigma^2$

## 3. 📊 Visual Understanding

Before we code, let's think about how these "knobs" work:

*   **Gaussian ($\mu, \sigma$):** Changing the mean ($\mu$) shifts the bell curve left or right. Changing the standard deviation ($\sigma$) squishes or stretches the curve.
*   **Binomial ($p$):** As the probability of success ($p$) increases, the peak of the discrete bars moves towards the right (higher number of successes).

## 4. 💻 Python Implementation

Here is how you generate and visualize all four in Python using `scipy.stats`.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, expon, norm

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Bernoulli (p=0.7) - e.g., 70% chance to pass a test
x_bern = [0, 1]
p_bern = 0.7
axes[0, 0].bar(x_bern, bernoulli.pmf(x_bern, p_bern), color='skyblue')
axes[0, 0].set_title('Bernoulli(p=0.7)')
axes[0, 0].set_xticks([0, 1])
axes[0, 0].set_xticklabels(['Fail (0)', 'Success (1)'])
axes[0, 0].set_ylabel('Probability')

# 2. Binomial (n=20, p=0.5) - e.g., 20 coin flips
n_binom, p_binom = 20, 0.5
x_binom = np.arange(0, n_binom + 1)
axes[0, 1].stem(x_binom, binom.pmf(x_binom, n_binom, p_binom))
axes[0, 1].set_title('Binomial(n=20, p=0.5)')
axes[0, 1].set_xlabel('Number of Successes')
axes[0, 1].set_ylabel('Probability')

# 3. Exponential (lambda=1.5) - e.g., time between server crashes
lam_exp = 1.5
x_exp = np.linspace(0, 5, 100)
# Note: scipy uses 'scale' which is 1/lambda
axes[1, 0].plot(x_exp, expon.pdf(x_exp, scale=1/lam_exp), color='orange', lw=2)
axes[1, 0].fill_between(x_exp, expon.pdf(x_exp, scale=1/lam_exp), alpha=0.3, color='orange')
axes[1, 0].set_title('Exponential(λ=1.5)')
axes[1, 0].set_xlabel('Time between events')
axes[1, 0].set_ylabel('Density')

# 4. Gaussian/Normal (mu=0, sigma=1) - Standard Normal
mu_norm, sig_norm = 0, 1
x_norm = np.linspace(-4, 4, 100)
axes[1, 1].plot(x_norm, norm.pdf(x_norm, mu_norm, sig_norm), color='green', lw=2)
axes[1, 1].fill_between(x_norm, norm.pdf(x_norm, mu_norm, sig_norm), alpha=0.3, color='green')
axes[1, 1].set_title('Gaussian(μ=0, σ=1)')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

## 5. 🔥 Interview + Exam Focus

### **The "Memoryless" Trick**
Exponential is the only continuous memoryless distribution.
*   **Interview question:** "If you've waited 10 minutes for a bus, is it more likely to arrive in the next minute?"
*   **Answer:** **NO.** If wait time is strictly exponential, the system has no "memory" of how long you've waited. The probability of waiting 1 more minute is the same as if you just arrived.

### **Why Gaussian Everywhere?**
Be prepared to mention the **Central Limit Theorem (CLT)**. It states that if you add up enough independent random variables (even if they aren't Gaussian themselves), their sum will approximate a Gaussian distribution. This is why human height, IQ scores, and measurement errors all look like bell curves.

## 6. ⚠️ Common Mistakes

1.  **Confusing Rate ($\lambda$) with Mean in Exponential:** If an exam says "an event happens 4 times per hour", $\lambda = 4$. But the mean waiting time is $1/4$ of an hour (15 minutes). Don't plug 4 in as the mean!
2.  **Bernoulli vs. Binomial:** Bernoulli is literally just a Binomial distribution where $n=1$.
